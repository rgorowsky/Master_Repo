```python
 # Dependencies
from bs4 import BeautifulSoup
import requests
import os
import pymongo
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.nasaPage_db
collection = db.items
```

```python
# URL of page to be scraped
url = 'https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')
```

```python
# THis is my first "MASS" search, to just see what's there
results = soup.find_all('div')
results
```

```python
# Now I'll see if I can pull out all of the titles, and paragraphs
specificResults = soup.find_all('div', class_='image_and_description_container')
specificResults
```

```python
# attempting to filter for features only 
filterResults = soup.find_all('div', class_='content_title')
filterResults

# class="content_title"
```

```python
# Loop through returned results
for result in filterResults:
    
    # Retrieve the thread title
    title = result.select_one('a')
    
    print(title)
    
    
    # Access the thread's text content
    title_text = title.a.text
    print(title_text)
```
